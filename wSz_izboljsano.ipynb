```python
def wSz(M):
    indeks = []
    d = M.distance_all_pairs()
    for u,v in M.edges(labels = false):
        blizu_u = 0
        for a in M.vertices():
            if d[a][u] < d[a][v]:
                blizu_u += 1
            blizu_v = order(M)- blizu_u
        indeks += [(M.degree(u) + M.degree(v)) * blizu_u * blizu_v]
    return sum(indeks)
```

```python
def novo_vozlisce(G):
    H = Graph(G)
    novo_vzl = order(H)
    H.add_edge((0,novo_vzl,None))
    return H
```

```python
H = Graph([(0, 1), (1, 2),(0,3),(3,4),(0,5),(5,6)])
novo_vozlisce(H)
```

```python
def min_wSz(H,koraki):
    k = 0
    primerjajH = H
    HwSz = wSz(H)
    while k < koraki:
        k = k + 1
        H = spremeni_graf(H)
        MwSz = wSz(H)
        if MwSz < HwSz:
            primerjajH = H
    return primerjajH
```
