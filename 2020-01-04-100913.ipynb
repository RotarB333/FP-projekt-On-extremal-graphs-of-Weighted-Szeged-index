```python
def wSz(M):
    indeks = 0
    for u,v in M.edges(labels = false):
        vsota = M.degree(u) + M.degree(v)
        blizu_u = 0
        blizu_v = 0
        for a in M.vertices():
            if M.distance(a,u)< M.distance(a,v):
                blizu_u += 1
            else:
                blizu_v += 1
        indeks += vsota * blizu_u * blizu_v
    return indeks
```

```python
def minimalni_wSz(a):
    M = []
    L = list(graphs(a))
    for i in range(0,len(L)):
        if L[i].is_connected() and L[i].is_tree:
            M += [L[i]]
    m1 = wSz(M[0])
    minM = M[0]
    for m in M:
        if wSz(m) < m1:
            m1 = wSz(m)
            minM = m
    return minM
```
