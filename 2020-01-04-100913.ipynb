```python
def wSz(M):
    indeks = []
    for u,v in M.edges(labels = false):
        blizu_u = 0
        for a in M.vertices():
            if M.distance(a,u)< M.distance(a,v):
                blizu_u += 1
            blizu_v = order(M)- blizu_u
        indeks += [(M.degree(u) + M.degree(v)) * blizu_u * blizu_v]
    return sum(indeks)
```

```python
def minimalni_wSz(G):
    novo = order(G)
    G.add_edge((0,novo,None))
    k = 0
    zacetniG = G
    while k < 10000:
        k += 1
        G = zacetniG
        M = graphs.CompleteGraph(novo + 1)
        for u,v,w in G.edges():
            M.delete_edge(u,v,w)
        G.delete_edge(G.random_edge())
        G.add_edge(M.random_edge())
        if (G.is_connected()and wSz(G)< wSz(zacetniG)):
                zacetniG = G
    return (zacetniG, wSz(zacetniG))

H = Graph([(0, 1), (1, 2),(0,3),(3,4),(0,5)])
N = minimalni_wSz(H)
(N[0].show(),N[1])
```

```python
def ustrezni_grafi(a):
    M = []
    L = list(graphs(a))
    for i in range(0,len(L)):
        if L[i].size() == a-1:
            if L[i].is_connected():
                M += [L[i]]
    return M

def min_wSz(M):
    m1 = wSz(M[0])
    minM = M[0]
    for m in M:
        if wSz(m) < m1:
            m1 = wSz(m)
            minM = m
    return minM
```
